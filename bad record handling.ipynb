```python
from pyspark.sql.types import StructField,StructType,StringType,IntegerType
schema=StructType([StructField("Month",StringType(),True),
                   StructField("Emp_count",IntegerType(),True),
                   StructField("Production_unit",IntegerType(),True),
                   StructField("Expense",IntegerType(),True),
                   StructField("_corrupt_record",StringType(),True)])
```

```python
df=spark.read.option("mode","permissive").option("header",'true').schema(schema).csv("/Volumes/workspace/bronze/circuit/image_table.csv")
display(df)
```

```python
df_malformed=spark.read.option("mode","dropmalformed").option('header','true').schema(schema).csv("/Volumes/workspace/bronze/circuit/image_table.csv")
display(df_malformed)
```

```python
df=spark.read.option('mode','failfast').option('header','true').option('schema',schema).csv("/Volumes/workspace/bronze/circuit/image_table.csv")
display(df)
```
